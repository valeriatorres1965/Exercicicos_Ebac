Lê o conteudo do arquivo em uma lista onde cada elemento é um dicionário representando as linhas do arquivo.

```python
emprestimos = []

with open(file='./credito.csv', mode='r', encoding='utf8') as fp:
  fp.readline()
  linha = fp.readline()
  while linha:
    linha_emprestimo = {}
    linha_elementos = linha.strip().split(sep=',')
    linha_emprestimo['id_vendedor'] = linha_elementos[0]
    linha_emprestimo['valor_emprestimos'] = linha_elementos[1]
    linha_emprestimo['quantidade_emprestimos'] = linha_elementos[2]
    linha_emprestimo['data'] = linha_elementos[3]
    emprestimos.append(linha_emprestimo)
    linha = fp.readline()
```

```python
for emprestimo in emprestimos:
  print(dict(emprestimo))
```

```python
valor_emprestimos_lista = []

with open(file='./credito.csv', mode='r', encoding='utf8') as fp:
  fp.readline()
  linha = fp.readline()
  while linha:
    linha_separada = linha.split(sep=',')
    valor_emprestimos = linha_separada[1]
    valor_emprestimos= float(valor_emprestimos)
    valor_emprestimos_lista.append(valor_emprestimos)
    linha = fp.readline()
  print(valor_emprestimos_lista)
```

```python
identificador_chave = [
  {'id_vendedor': 104271, 'valor_emprestimos': 448.0},
  {'id_vendedor': 21476, 'valor_emprestimos': 826.7},
  {'id_vendedor': 87440, 'valor_emprestimos': 313.6},
  {'id_vendedor': 15980, 'valor_emprestimos': -8008.0},
  {'id_vendedor': 215906, 'valor_emprestimos': 2212.0},
  {'id_vendedor': 33696, 'valor_emprestimos': 2771.3},
  {'id_vendedor': 33893, 'valor_emprestimos': 2240.0},
  {'id_vendedor': 123974, 'valor_emprestimos': 2021.95},
  {'id_vendedor': 225870, 'valor_emprestimos': 4039.0}

]
```

```python
def chave_ident_emprestimo(identificador_chave, new_valor_empr):
  def apply(x):
    x['id_vendedor'] = new_valor_empr
    return x
  return map(apply, identificador_chave)
```

```python
chave_ident_emprestimos = chave_ident_emprestimo(identificador_chave, 104271)
```

```python
print(list(chave_ident_emprestimos))
```

```python
from functools import reduce

soma_valor_emprestimo = valor_emprestimos_lista
print(reduce(lambda x, y: x+y, soma_valor_emprestimo))
```
